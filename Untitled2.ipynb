```python
# Імпорт бібліотек для роботи з даними
import pandas as pd
```

```python
# Читаємо CSV файл у DataFrame
df = pd.read_csv('dataset_2026-04-19T22_59_10.915577463Z_DEFAULT_INTEGRATION_SSSU_DF_POPULATION_BIRTH_LATEST.csv')
```

```python
# Перегляд структури даних
df.columns.tolist()
```

```python
## Фільтрація даних
df.columns = df.columns.str.strip()
```

```python
# Залишаємо тільки дані по Україні.
# На цьому етапі ми залишили лише записи, що стосуються України, щоб аналіз був точним і локальним.

df = df[df['Територіальний розріз'] == 'Україна']
```

```python
# 📌 Обираємо роки 2014–2021
year_cols = [str(y) for y in range(2014, 2022)]
year_cols = [col for col in year_cols if col in df.columns]
```

```python
# Перетворення таблиці у довгий формат (long format)
id_vars = [
    'Показник',
    'Сімейний cтан',
    'Громадянство матері',
    'Громадянство батька'
]
```

```python
df_long = df.melt(
    id_vars=id_vars,
    value_vars=year_cols,
    var_name='year',
    value_name='value'
)
```

```python
# Видалення пропусків і конвертація в числа
df_long = df_long.dropna()
df_long['value'] = pd.to_numeric(df_long['value'], errors='coerce')
df_long = df_long.dropna()
```

```python
# Фільтруємо тільки народжуваність
df_births = df_long[df_long['Показник'].str.contains('народж', case=False)]
```

```python
# Розрахунок сумарної народжуваності по роках
df_yearly = df_births.groupby('year', as_index=False)['value'].sum()
df_yearly
```

```python
# Імпорт даних
import matplotlib.pyplot as plt
#Лінійний графік
# Динаміка народжуваності
## Графік показує стабільне зниження народжуваності в Укаїні.
plt.plot(df_yearly['year'], df_yearly['value'])
plt.title('Народжуваність в Україні (2014–2021)')
plt.xlabel('Рік')
plt.ylabel('Кількість народжень')
plt.show()
```

```python
#Річні зміни
df_yearly['growth_%'] = df_yearly['value'].pct_change() * 100
```

```python
#Графік показує,як змінюється народжуваність рік до року у %.
plt.plot(df_yearly['year'], df_yearly['growth_%'])
plt.title('Зміна народжуваності (% рік до року)')
plt.xlabel('Рік')
plt.ylabel('% зміни')
plt.show()
```

```python
#Стовпчиковий графік
## Порівняння по роках дає можливість наочно побачити спад показника
plt.bar(df_yearly['year'], df_yearly['value'])
plt.title('Народжуваність в Україні (2014–2021)')
plt.xlabel('Рік')
plt.ylabel('Кількість народжень')
plt.show()
```
